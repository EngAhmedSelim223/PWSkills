##### Q1. Explain GET and POST methods.

```python
# Q1.Answer:

# GET method --> is used to submit the HTML form data. This data is visible to everyone in the browser's address bar, 
# which means that all the variable names and their values will be displayed in the URL and this not good for the security

# POST method -->  is also used to submit the HTML form data but unlike GET Unlike the GET method, it does not have a limit on the amount of information to be sent.
# The information sent from an HTML form using the POST method is not visible to anyone.
```
