##### Q2. Write a python program to print Python Exception Hierarchy.

```python
# Q2.Answer:

def hierarcy(cls, v = 0):
    print("_" * v, cls.__name__)
    for i in cls.__subclasses__():
        hierarcy(i, v + 3)
        
print("Hierarchy for Built-in exceptions is : ")
hierarcy(BaseException)
```

##### Q4. Why LookupError class is used? Explain with an example KeyError and IndexError.

```python
# Q4.Answer:

# Because Lookup error is the base class for errors that occurs when we do not found something
try:
    dct = {'name':"ahmed selim",'age':28  }
    print(dct['email'])
except KeyError as e:
    print("there is an issue as : ", e, "key not found")
```

```python
try:
    lst = [1,2,3,4]
    print(lst[5])
except IndexError as e :
    print("there is an issue as : ", e)
```
