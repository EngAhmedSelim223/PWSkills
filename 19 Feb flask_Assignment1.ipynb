##### Q5. What function is used in Flask for URL Building? Write a Python code to demonstrate the working of the url_for() function.

```python
# Q5.Answer: 

# We use url_for() to build and generate URLs on a Flask application in a dynamic way 
from flask import Flask, redirect, url_for
app = Flask(__name__)

@app.route("/account_type/<ac_type>")
def account_type(ac_type):
    return f"this is {ac_type} account"


@app.route("/user/<name>")
def user(name):
    if name == "ahmed":
        return redirect(url_for('account_type', ac_type = "Premium"))
    else:
        return redirect(url_for("account_type", ac_type = "FREE"))
    
if __name__ == "__main__":
    app.run(host="0.0.0.0")
```
